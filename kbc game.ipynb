```python
import time
import random
import pygame

pygame.init()
def bg_music():
    global music1
    music1 = pygame.mixer.Sound(r"C:\Users\Lenovo\Downloads\bgmusic.mp3")
    music1.play()
    
def correct_sound():
    global music
    music = pygame.mixer.Sound(r"D:\c-downloads\correct.mp3")
    music.play()
    
def wrong_sound():
    global music
    music = pygame.mixer.Sound(r"D:\c-downloads\wrong.mp3")
    music.play()
    
def crore_sound():
    global music
    music = pygame.mixer.Sound(r"C:\Users\Lenovo\Downloads\crore.mp3")
    music.play()


questions = ['Which one of the following river flows between Vindhyan and Satpura ranges?',
            'The Central Rice Research Station is situated in?',
            'Who among the following wrote Sanskrit grammar?',
            'The metal whose salts are sensitive to light is?']

options = [['Narmada','Mahanadi','Son','Netravati'],
          ['Chennai','Cuttack','Bangalore','Quilon'],
          [ 'Kalidasa','Charak','Panini','Aryabhatt'],
          ['Zinc','Silver','Copper','Aluminum']]

correct = ['Narmada','Cuttack','Panini','Silver']


x = 'ABCD'

bg_music()

print('Welcome to KBC!!')
for i in range(len(questions)):
    win = False
    print(f'Question No: {i+1}')
    print(questions[i])
    for index,value in enumerate(options[i]):
        print(f'{x[index]}-> {value}')
    print()
    
    answer = input('Enter your answer: ')
    
    if answer.lower() == correct[i].lower():
        print('Correct Answer üëç')
        correct_sound()        
        win = True
    
    else:
        print('Wrong Answer üëé')
        music1.stop()
        wrong_sound()
        break
    
    print()

if win == True:
    print('You Won The Game!!')
    music1.stop()
    crore_sound()  
else:
    print('You loss The Game!!')
```
