```python
meesho_dict = {'Sunglasses Name':[],
              'price':[]}


for i in soup.find_all('div',{'class':'NewProductCardstyled__ProductHeaderWrapper-sc-6y2tys-30 gspQJ'}):
    name,price = i.text.strip().split('₹')
    price  = price.split(' ')[1]
    
    meesho_dict['Sunglasses Name'].append(name.strip())
    meesho_dict['price'].append(price)
    
print('done')
```
